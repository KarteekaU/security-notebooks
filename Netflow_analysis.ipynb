```python
# labeling -- 1's if has 'botnet', o's otherwise

is_botnet = df['Label'].apply(lambda x: 'Botnet' in x)
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(random_state = 0).fit(df[['TotBytes']], is_botnet)
```
